I am going to look at the quality values and create my cutoff point for high and low quality wines. This way we have a binary target column (high quality wines) that we can look for.

```python
wine_table['quality'].value_counts()
```

```python
# I'm going to write a function specifically for splitting the wine quality because values of 6 are very high
# and I want to see if changing high quality wines to include/exclude ratings of 6 have any influence on predictions
def wine_quality_split(table, cutoff):
    # cutoff is included in low quality
    table['h_quality'] = table.apply(lambda row: 1 if row.quality > cutoff else 0, axis=1)
    table['l_quality'] = table.apply(lambda row: 0 if row.h_quality == 1 else 1, axis=1)
    return table
```

```python
# going to set the cutoff at 6 so the high quality counts should only include 7 and 8
wine_table = wine_quality_split(wine_table, 6)

# the value counts should be opposite
print(wine_table['h_quality'].value_counts())
print()
print(wine_table['l_quality'].value_counts())
```

I am going to use K-Folding to test the decision tree and see what the best values of k are for modeling this data. 

```python
# none of these columns are binary so I am going to OHE some of them for predictors
# I am going to use chlorides, total sulfur dioxide, residual sugar, density, and quality

ohe_columns = [
    'chlorides',
    'total sulfur dioxide',
    'residual sugar',
    'density',
    'quality'
]
```

```python
def batch_ohe(table, targets, bins):
    """ writing this function to One Hot Encode multiple non-binary columns in one function call"""
    for target in targets:
        print("Binning " + target)
        bin_names = []
        bin_names.append('Low')
        bin_names.append('Medium')
        bin_names.append('High')
        
        bin_name = target.replace(' ', '') + '_bin'
        
        table[bin_name] = pd.cut(table[target], bins, labels=bin_names)
        print(table[bin_name].value_counts())
        print("One Hot Encoding " + target)
        
        target = target.replace(' ', '')
        one_hot_e = pd.get_dummies(table[bin_name],prefix=target,dummy_na=False)  # false because should not have empties
        table = table.join(one_hot_e)
        
        print()
        
        table = table.drop(bin_name, axis=1)
    return table
```

```python
wine_table = batch_ohe(wine_table, ohe_columns, 3)
```

```python
#pulled from my K_Means notebook.

def assign_nulls(table, target):
    """
    table => our input table (wine_table in this case)
    target => our target column
    """
    #length of column
    column_size = len(table[target])
    
    #using a ratio of 10 to 30 percent nulls, may add customization to this later
    ratio = float(random.randint(10, 30)/100)
    int_ratio = int(ratio * column_size)
    
    #gather random indices
    indices = random.sample(range(column_size), int_ratio)
    #sourced from https://stackoverflow.com/questions/6482889/get-random-sample-from-list-while-maintaining-ordering-of-items
    
    #assign nulls to randomly selected indices
    for i in indices:
        table.at[i, target] = float("NaN")
    
    return table
```

```python
#assign random nulls to all my ohe columns
wine_table['quality'] = wine_table.apply(lambda row: float(row.quality), axis=1)
for target in ohe_columns:
    wine_table=assign_nulls(wine_table, target)
```

# Reflection

Data Wrangling:
This project started out a bit rocky because I didn't think originally I would use NaN values in the data for this portion. This proved to be very important for the accuracy of the models later. Also the cutoff for quality was super important. That's what the first third of this notebook is about. On the topic of the NaN values I learned that without these, the algorithms would accurately model every time because there was were predictions going on in the data, all they had to do was read it. Next the cutoff helped with this as well. Since the quality values of 5 and 6 are roughly equal in size in the wine data set having them both reserved as low quality divides the data set very cleanly. Anything that is classified as high quality could be (aggresively) argued as an outlier. Including 6 in the high quality basket mixed the bag up because the stats of wines rated at 6 and the stats at 5 are very close together. It became very difficult for k folding and random forests to find a defining factor between high and low quality. 

K folding and random forests:
All in all I think that k folding creates a better model than the random forests for two reasons. One, the random forest doesn't improve over an increase in the amount of trees. There seems to be a sweet spot of accuracy around 15-20 trees but then the tests decrease in accuracy with lower and higher k values. So the first reason is that the random forest has a cap on it's accuracy and the cap is not great to begin with. The second reason has to do with the comparison of the algorithms themselves. When looking at the value counts between random forests and k folding, the random forests rarely have false_positives and their performance of true_negatives and false_negatives are very high. What is happening here is that the random forests are classifying a lot of points wrong. The random forest is having trouble distinguishing between the cutoff 5-6 and usually choosing the incorrect classification. This does not improve with more data. The random forest seems to get confused with too little and too much data. K folding handles this growth of information much better. As the k value gets pushed higher and higher the accuracy of the model consistently improves. The only downfall is when k folding is not given enough folds.
