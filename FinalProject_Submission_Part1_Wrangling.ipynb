```python
import random

#assign nulls
def assign_nulls(table, target):
    """
    table => our input table (wine_table in this case)
    target => our target column
    """
    #length of column
    column_size = len(table[target])
    
    #using a ratio of 5 to 20 percent nulls, may add customization to this later
    ratio = float(random.randint(5, 20)/100)
    int_ratio = int(ratio * column_size)
    
    #gather random indices
    indices = random.sample(range(column_size), int_ratio)
    #sourced from https://stackoverflow.com/questions/6482889/get-random-sample-from-list-while-maintaining-ordering-of-items
    
    #assign nulls to randomly selected indices
    for i in indices:
        table.at[i, target] = "NaN"
    
    return table
```

```python
wine_table['prev_density'] = wine_table['density']
wine_table = assign_nulls(wine_table, 'density')
wine_table['no_density'] = wine_table.apply(lambda row: 1 if pd.isnull(row.density) else 0, axis=1)
```

Now I have to compute centroids for all the rows that have a 'Density' and all the rows that have 'Quality'. Add the respective values to each centroid and compute the average of the final, that way I have an average of each centroid. 

These isnull() checks may have counts that vary from the isnull() checks above because some rows may have NaN for density and quality and are removed subsequently.
