I am going to look at the quality values and create my cutoff point for high and low quality wines. This way we have a binary target column (high quality wines) that we can look for.

```python
# I'm going to write a function specifically for splitting the wine quality because values of 6 are very high
# and I want to see if changing high quality wines to include/exclude ratings of 6 have any influence on predictions
def wine_quality_split(table, cutoff):
    # cutoff is included in low quality
    table['h_quality'] = table.apply(lambda row: 1 if row.quality > cutoff else 0, axis=1)
    table['l_quality'] = table.apply(lambda row: 0 if row.h_quality == 1 else 1, axis=1)
    return table
```

```python
# going to set the cutoff at 6 so the high quality counts should only include 7 and 8
wine_table = wine_quality_split(wine_table, 5)

# the value counts should be opposite
print(wine_table['h_quality'].value_counts())
print()
print(wine_table['l_quality'].value_counts())
```

```python
# none of these columns are binary so I am going to OHE some of them for predictors
# I am going to use chlorides, total sulfur dioxide, residual sugar, density, and quality

ohe_columns = [
    'chlorides',
    'total sulfur dioxide',
    'residual sugar',
    'density',
    'quality'
]
```

```python
def batch_ohe(table, targets, bins):
    """ writing this function to One Hot Encode multiple non-binary columns in one function call"""
    for target in targets:
        print("Binning " + target)
        bin_names = []
        bin_names.append('Low')
        bin_names.append('Medium')
        bin_names.append('High')
        
        bin_name = target.replace(' ', '') + '_bin'
        
        table[bin_name] = pd.cut(table[target], bins, labels=bin_names)
        print(table[bin_name].value_counts())
        print("One Hot Encoding " + target)
        
        target = target.replace(' ', '')
        one_hot_e = pd.get_dummies(table[bin_name],prefix=target,dummy_na=False)  # false because should not have empties
        table = table.join(one_hot_e)
        
        print()
        
        table = table.drop(bin_name, axis=1)
    return table
```

```python
splitter_columns = [
    'chlorides_Low',
    'chlorides_Medium',
    'chlorides_High',
    'residualsugar_Low',
    'residualsugar_Medium',
    'residualsugar_High',
    'density_Low',
    'density_Medium',
    'density_High',
    'quality_Low',
    'quality_Medium',
    'quality_High'
]
```

```python
#pulled from my K_Means notebook.

def assign_nulls(table, target):
    """
    table => our input table (wine_table in this case)
    target => our target column
    """
    #length of column
    column_size = len(table[target])
    
    #using a ratio of 10 to 30 percent nulls, may add customization to this later
    ratio = float(random.randint(10, 30)/100)
    int_ratio = int(ratio * column_size)
    
    #gather random indices
    indices = random.sample(range(column_size), int_ratio)
    #sourced from https://stackoverflow.com/questions/6482889/get-random-sample-from-list-while-maintaining-ordering-of-items
    
    #assign nulls to randomly selected indices
    for i in indices:
        table.at[i, target] = float("NaN")
    
    return table
```

```python
#assign random nulls to all my ohe columns
wine_table['quality'] = wine_table.apply(lambda row: float(row.quality), axis=1)
for target in ohe_columns:
    wine_table=assign_nulls(wine_table, target)
```

# Reflection

Alright, this did not go as planned. There was only 1 sequence of votes that belonged to each rating exclusively. Upon realizing this I tried to identify trends in the placement and counts of 1 and 0 in each classification that would help clearly identify 5 or 6. This proved very difficult because of the roughly 50/50 ratio of vote counts between ratings of 5 and 6.

For example, when I ran this the first time the row with predictions "001100011000100" had a count of 262 in wine ratings of 5 and 220 in wine ratings of 6. Since we know that a wine rating of 6 is more likely to be incorrectly classified, we can start to identify that predictions of this count in context of a wine rating of 6 are actually high quality. This high amount of matches is bad when it comes to distinguishing high and low quality. 

What was interesting is I found there is only one row that distinctly does not match in each classifier. 

This will be very difficult to solve because the classification of 5 vs 6 are nearly identical and have very similar quantities of those classifications. Like I said before 262 5's and 220 6's have the classification '001100011000100' meaning that there is about a 50% chance the random forest gets this right. 

I can think of two ways to solve this but it would have to happen at the wrangling period. One, I could one hot encode another column (which would create 3) to include in the random forest and see if that extra value differentiates enough between quality ratings of 5 and 6. Two, adding stricter boundaries when one hot encoding the already existing columns, specifically when they are binned. Hopefully in one of the columns there is a numerical difference where for example "6's always have a lower acidity" but there is a possibility that such a limit does not exist. 

By looking at the matrix of predictions I have verified that the issue with classification of quality is distinguishing between ratings of 5 and 6 because they overwhelmingly get the same votes. This means that when conducting a study on this data in terms of quality that 6 should be a cutoff for high quality.
